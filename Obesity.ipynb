```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.cluster import KMeans
from sklearn.svm import SVC
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import cross_val_score
le=LabelEncoder()
train=pd.read_csv('train(1).csv')
test=pd.read_csv('test(1).csv')
x=train.drop(['NObeyesdad','id'], axis='columns')
xt=test.drop(['id'], axis='columns')
y=pd.DataFrame(train['NObeyesdad'])
def isnu(ex):
    for x in ex.isna():
        if x == True:
            print(0)
            break
abc=pd.concat([x['CALC'],xt['CALC']], axis=0, ignore_index=True )
id=pd.DataFrame(test['id'])
id
```

```python
x['Gender']=le.fit_transform(x['Gender'])
x['family_history_with_overweight']=le.fit_transform(x['family_history_with_overweight'])
x['FAVC']=le.fit_transform(x['FAVC'])
x['CAEC']=le.fit_transform(x['CAEC'])
x['SMOKE']=le.fit_transform(x['SMOKE'])
x['SCC']=le.fit_transform(x['SCC'])
le.fit(abc)
x['CALC']=le.transform(x['CALC'])
x['MTRANS']=le.fit_transform(x['MTRANS'])
x
```

```python
le.fit(train['Gender'])
xt['Gender']=le.transform(xt['Gender'])

le.fit(train['family_history_with_overweight'])
xt['family_history_with_overweight']=le.transform(xt['family_history_with_overweight'])

le.fit(train['FAVC'])
xt['FAVC']=le.transform(xt['FAVC'])

le.fit(train['CAEC'])
xt['CAEC']=le.transform(xt['CAEC'])

le.fit(train['SMOKE'])
xt['SMOKE']=le.transform(xt['SMOKE'])

le.fit(train['SCC'])
xt['SCC']=le.transform(xt['SCC'])

le.fit(abc)
xt['CALC']=le.transform(xt['CALC'])

le.fit(train['MTRANS'])
xt['MTRANS']=le.transform(xt['MTRANS'])

y=le.fit_transform(y)
y
```

```python
rfc=SVC(C=10,gamma=20)
rfc.fit(x,y)
rfc.score(x,y)
```

```python
predicted=rfc.predict(xt)
bestp=le.inverse_transform(predicted)
curp=pd.Series(bestp, name='NObeyesdad')
cp=pd.DataFrame(curp)
po=pd.concat([id,cp],axis='columns')
po.to_csv('final.csv',index=False)
```
